```python
import pandas as pd

url = "https://bit.ly/4nejNue"
data = pd.read_csv(url)
```

```python
# Show the first few rows
data.head()

# Basic statistics
data.describe()

# Mean, Median, Std
print("Mean:\n", data.mean(numeric_only=True))
print("\nMedian:\n", data.median(numeric_only=True))
print("\nStandard Deviation:\n", data.std(numeric_only=True))
```

```python
import matplotlib.pyplot as plt

# Plot histograms of all numeric columns
data.hist(figsize=(10, 8), bins=20, edgecolor='black')
plt.suptitle("Histograms of All Features", fontsize=16)
plt.tight_layout()
plt.show()
```
